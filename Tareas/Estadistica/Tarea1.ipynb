```python
ciclos=10000
a=1
N=int(input('Ingrese el numero de pasos:'))
pos_final=zeros(ciclos)

def caminata(numPasos,paso):
    pos=0
    for i in range(0,numPasos):
        p=random.random()
        if(p>=0.5):
            pos+=paso
        else:
            pos-=paso
    return pos

for i in range(0,ciclos):
        pos_final[i]=caminata(N,a)

    
        
```

```python
hist(pos_final, bins=41)
title('Distribución de posiciones')
```

```python
Ns=list(range(1000, 51000, 1000))
n=100
Arrmediax = zeros(len(Ns))
Arrmediax2 = zeros(len(Ns))

for i in range(0,50):
    mediax = 0
    mediax2 = 0
    
    for j in range (0,n):
        x=caminata(i,1)
        mediax += x*(1.0/n)
        mediax2 += (x**2)*(1.0/n)
    Arrmediax[i] = mediax
    Arrmediax2[i] = mediax2
```

```python
plot(Ns, Arrmediax, label='<x>')
plot(Ns, Arrmediax2, label='<x2>')
xlabel('N')
legend(loc='upper left')
```
