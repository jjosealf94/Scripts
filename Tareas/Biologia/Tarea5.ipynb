##Anexo 1

```python
%pylab inline
co=10
x=linspace(0,5,50) 
figure(1)

subplot(2,1,1)
title('Ecuaciones dinámicas de a(t), i(t)')
plot(x,co*(2-exp(-5*x)),'-r',label='a(t)')
plot(x,co*(2-exp(-x)),'-b', label='i(t)')
xlim(-0.2,5)
ylim(10,22)
legend(loc="center right") 
show()

subplot(2,1,2)
title('Ecuación dinámica de rss(t) adaptación en [] uniforme')
plot(x,(2-exp(-5*x))/(2-exp(-x)),'-g',label='rss(t)')
xlim(-0.2,5)
legend(loc="center right")
show()
```

##Anexo 2

```python
co=1
c1=1
DiYi=10 #Cociente entre Di/Yi >> 1
x=linspace(0,2*pi,100) 
title('Ecuación dinámica rss($\\theta$) polarizada en tetha')
plot(x,(co+(c1*cos(x)))/(co+((c1/DiYi)*cos(x))),'-r',label='rss($\\theta$)')
legend(loc="upper right") 
show()
```

```python
x=linspace(0,2*pi,100) 
figure(3)
title('concentración c($\\theta$)= co+c1cos($\\theta$) (Azul) y rss($\\theta$) (rojo) DY=10>>1')
plot(x,co+c1*cos(x), '-b')
plot(x,(co+(c1*cos(x)))/(co+((c1/DiYi)*cos(x))),'-r')
show()
```
