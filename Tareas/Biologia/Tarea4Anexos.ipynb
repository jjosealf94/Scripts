###<B><FONT COLOR="red">Anexo 1c </FONT>

```python
Pk=5.0/15.0 #Probabilidad de que ocurra reacción B
Pr=10.0/15.0 #Probabilidad de que ocurra reacción A
contr=0 #Contador RA
contk=0 #Contador RB
tr=[] #tiempo RA
tk=[] #tiempo RB
tt=[] #tiempo total

t=0.0
while (contr<50) | (contk<50): # Tiene que ser un o porque quiero que ambos sean mayor 50 eventos
    a=random.random()
    if (a<Pr): # Si el # aleatorio es menor a 2/3 ocurre A pero NO ocurre B
        contr+=1
        tr.append(t)
        tt.append(t)
    elif(a>Pr): # Si el # aleatorio es mayor a 2/3 NO ocurre A pero ocurre B
        contk+=1
        tk.append(t)
        tt.append(t)
    b=random.exponential(1.0/15.0)
    t=t+b

tr=array(tr)
Deltatr=zeros(len(tr)-1)

tk=array(tk)
Deltatk=zeros(len(tk)-1)

tt=array(tt)
Deltatt=zeros(len(tt)-1)


for i in range (0,len(tr)-1):
    Deltatr[i]=tr[i+1]-tr[i]

for i in range (0,len(tk)-1):
    Deltatk[i]=tk[i+1]-tk[i]
    
for i in range (0,len(tt)-1):
    Deltatt[i]=tt[i+1]-tt[i]

    
print("El número de eventos reacción A",contr)
print("El número de eventos reacción B",contk)
```

```python
def expo(x, a, b):
    return a*exp(-b*x)
```

###<B><FONT COLOR="red">Nota:Para los cálculos de los promedios de las distribuciones exponenciales ver nota dejada en el punto 2c </FONT>

```python
figure(1)
count, bins, ignored = hist(Deltatr, bins=40,normed=True)
title('Histograma de tiempos para reacción A',size=15)
popt, pcov = curve_fit(expo, bins[:40], count)
plot(bins,expo(bins,popt[0],popt[1]), color='red')
xlim(0,0.4)
print("el promedio del fit de Reacción A es=",popt[0]/popt[1]**2)

figure(2)
count, bins, ignored = hist(Deltatk, bins=30,normed=True)
title('Histograma de tiempos para reacción B',size=15)
popt, pcov = curve_fit(expo, bins[:30], count)
plot(bins,expo(bins,popt[0],popt[1]), color='red')
xlim(0,0.6)
print("el promedio del fit de Reacción B es=",popt[0]/popt[1]**2)


figure(3)
count, bins, ignored = hist(Deltatt, bins=35,normed=True)
title('Histograma de tiempos para el evento en general',size=15)
popt, pcov = curve_fit(expo, bins[:35], count)
plot(bins,expo(bins,popt[0],popt[1]), color='red')
xlim(0,0.3)
print("el promedio del fit del evento general es=",popt[0]/popt[1]**2)
```

###<B><FONT COLOR="red">Anexo 1d</FONT>

```python
Pa=15.0/15.0 #Siempre ocurre A, mientras no cambie a B (con prob 1/3)
conta=0 #Contador RA
contb=0 #Contador RB
ta=[] #tiempo RA
tb=[] #tiempo RB
tt=[] #tiempo total

t=0.0
while (conta<50) | (contb<50):
    a=random.random()
    if(a<1/3):
        contb+=1
        tb.append(t)
        tt.append(t)
    else:
        conta+=1
        ta.append(t)
        tt.append(t)
    b=random.exponential(1/15)
    t=t+b
    
ta=array(ta)
Deltata=zeros(len(ta)-1)

tb=array(tb)
Deltatb=zeros(len(tb)-1)

tt=array(tt)
Deltatt=zeros(len(tt)-1)


for i in range (0,len(ta)-1):
    Deltata[i]=ta[i+1]-ta[i]

for i in range (0,len(tb)-1):
    Deltatb[i]=tb[i+1]-tb[i]
    
for i in range (0,len(tt)-1):
    Deltatt[i]=tt[i+1]-tt[i]

    
print("El número de eventos reacción A",conta)
print("El número de eventos reacción B",contb)
```

```python
figure(1)
count, bins, ignored = hist(Deltata, bins=45,normed=True)
title('Histograma de tiempos para reacción A')
popt, pcov = curve_fit(expo, bins[:45], count)
plot(bins,expo(bins,popt[0],popt[1]), color='red')
xlim(0,0.4)
print("el promedio del fit de RA es=",popt[0]/popt[1]**2)

figure(2)
count, bins, ignored = hist(Deltatb, bins=35,normed=True)
title('Histograma de tiempos para reacción B')
popt, pcov = curve_fit(expo, bins[:35], count)
plot(bins,expo(bins,popt[0],popt[1]), color='red')
xlim(0,0.6)
print("el promedio del fit de RB es=",popt[0]/popt[1]**2)


figure(3)
count, bins, ignored = hist(Deltatt, bins=35,normed=True)
title('Histograma de tiempos para el evento en general')
popt, pcov = curve_fit(expo, bins[:35], count)
plot(bins,expo(bins,popt[0],popt[1]), color='red')
xlim(0,0.3)
print("el promedio del fit del evento general es=",popt[0]/popt[1]**2)
```
