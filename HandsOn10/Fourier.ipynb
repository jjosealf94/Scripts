```python
%pylab inline
from scipy.fftpack import fft, fftfreq
```

```python
#importa los datos a trabajar
data=genfromtxt("./monthrg.dat")
```

```python
year=data[:,0] + data[:,1]/12
promanchas=data[:,3]
year19=year[year>=1900]
promanchas19=promanchas[year>=1900]


fourierManchas=(fft(promanchas19))

N = len(promanchas19) 
f = 1./100 
dt = 1./f / N 
t = np.linspace(0,1/f,N)
freq = fftfreq(N, dt)


plot(freq,abs(fourierManchas))   
xlim(-0.5,0.5)
```

```python
fourierManchas[np.abs(freq)>0.2] = 0
y_cleaned=ifft(fourierManchas)

figure(1,figsize(10,7))
plot(year19,promanchas19,label='Ciclo sin filtro')
plot(year19,abs(y_cleaned),'r',lw=2,label='Filtrado')
title('manchas solares promedio mensual contra años')
legend()
show()
```
