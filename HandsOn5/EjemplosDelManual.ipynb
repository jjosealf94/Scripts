#Ejemplos del Manual de Matplotlib


Si se proporciona una sola lista al comando **plot()**, matplotlib asume que es la secuencia de los valores de Y de manera automática genera una lista para valores de x. Empezando en 0 y terminando con el mismo número de elementos.

```python
plt.plot([1,2,3,4])
plt.show()
```

Por cada pareja (x,y), hay un tercer argumento opcional el cuál es el color y el tipo de línea.
**Tipo de línea**: ’-’  solid line style, ’--’ dashed line style, ’-.’ dash-dot line style, ’o’ circle marker, ’^’ triangle_up marker.
**Color**: b (blue), g (green), c (cyan), r (red) , m (magneta), y (yellow), k (black), w (white)

El comando **axis()** toma una lista de [xmin, xmax, ymin, ymax] y ajusta los ejes a los nuevos valores.


```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,6,0,20])
plt.show()
```

matplotlib soporta el uso de **arreglos** de numpy, de hecho todas todas las secuencias son convertidas a arreglos internamente.

```python
import numpy as np 
import matplotlib.pyplot as plt
t=np.arange(0.,5.,0.2)
plt.plot (t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

Existen muchos atributos que se pueden modificar: Grosor de la línea, estilo, etiquetas, etc. Se puede usar el comando **setp()** para modificar los atributos de líneas. 
Pyplot trabaja sobre la actual figura y ejes y todos los comandos son aplicados a ellos. Se puede crear multiples figuras usando el comando **figure()** con un número que aumenta. Cada figura puede contener el número de ejes y 'subplots' que uno quiera. El comando **subplot()** especifica el número de filas, columnas y número de la figura (1-col*fil)

Finalmente el comando **axes()** permite establecer en que posición se quiere colocar los ejes. axes([left,bottom, width, height]), donde todos los valores van de 0 a 1.

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(2,1,1)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(2,1,2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

El comando **text()** puede ser usado para agregar texto en cualquier ubicación, **xlabel()**, **ylabel()** y **title()** para agregar texto en las posiciones indicadas. Todos los comandos de text() reciben dentro 'keywords' que personalizan el texto. Matplotlib aceta texto en forma de TeX para escribir ecuaciones, de la forma title(r'$\sigma_i=15$'). La **r** es importante ya que significa que en el string los backslashes sean tratados como tal y no como secuencias escape en python.

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

A veces es necesario poner el texto en una posición sobre la gráfica, para lo cual es útil el uso del comando **annotate()** ya que hace anotaciones de manera fácil. En una anotación hay dos puntos importantes a considerar: **La ubicación de la nota** representada por una pareja (x,y) y la **ubicación del texto** representada por otra pareja (x,y). 

```python
ax = plt.subplot(111)
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
arrowprops=dict(facecolor='black', shrink=0.05),
)
plt.ylim(-2,2)
plt.show()
```

**Ejemplos del uso de comandos de texto**

```python
fig = plt.figure()
fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)
ax.set_title('axes title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.text(3, 8, 'boxed italics text in data coords', style='italic',
bbox={'facecolor':'red', 'alpha':0.5, 'pad':10})
ax.text(2, 6, r'an equation: $E=mc^2$', fontsize=15)
#ax.text(3, 2, unicode('unicode: Institut f\374r Festk\366rperphysik', 'latin-1'))
ax.text(0.95, 0.01, 'colored text in axes coords',
verticalalignment='bottom', horizontalalignment='right',
transform=ax.transAxes,
color='green', fontsize=15)
ax.plot([2], [1], 'o')
ax.annotate('annotate', xy=(2, 1), xytext=(3, 4),
arrowprops=dict(facecolor='black', shrink=0.05))
ax.axis([0, 10, 0, 10])
plt.show()
```

Se pueden usar **TeX markup** en cualquier comando de texto de matplotlib colocando el texto dentro del signo dolar ($). No es necesario tener instalados TeX, dado que matplotlib tiene su propio analizador TeX de expresiones.

* Para realizar **subscripts and superscripts**, use the ’_’ and ’^’ symbols. 
* Para hacer fracciones, binomios y números apilados con los siguientes comandos \frac{}{}, \binom{}{} y \stackrel{}{} respectivamente.
* Para realizar raíces \sqrt[]{}
* Para personalizar los estilos \mathrm{Roman} \mathit{Italic} \mathtt{Typewriter} \mathcal{CALLIGRAPHY} 
* Para colocar acentos (\bar a) (\ddot a or \"a)  (\dot a or \.a)  (\hat a or \^a) (\vec a) (\overline{abc} a)



```python
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)
plt.plot(t,s)
plt.title(r'$\alpha_i > \beta_i$', fontsize=20)
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20)
plt.text(0.6, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',
fontsize=20)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```

La posición de la legenda puede ser especificada en el argumento con la 'keyword' loc en el comando **lengend()**. La 'keyword' bbox_to_anchor brinda un grado de control manual en la ubicación de la legenda.

```python
plt.subplot(2,1,1)
plt.plot([1,2,3], label="test1")
plt.plot([3,2,1], label="test2")
# Place a legend above this legend, expanding itself to
# fully use the given bounding box.
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,ncol=2, mode="expand", borderaxespad=0.)

plt.subplot(2,2,3)
plt.plot([1,2,3], label="test1")
plt.plot([3,2,1], label="test2")
# Place a legend to the right of this smaller figure.
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

Aveces es más claro mostrar multiples legendas dentro de la figura. Para lograr esto se usa la llamada al comando legend() las veces necesarias y usado la 'keyword' handles=[]

```python
line1, = plt.plot([1,2,3], label="Line 1", linestyle='--')
line2, = plt.plot([3,2,1], label="Line 2", linewidth=4)
# Create a legend for the first line.
first_legend = plt.legend(handles=[line1], loc=1)
# Add the legend manually to the current Axes.
ax = plt.gca().add_artist(first_legend)
# Create another legend for the second line.
plt.legend(handles=[line2], loc=4)
plt.show()
```
