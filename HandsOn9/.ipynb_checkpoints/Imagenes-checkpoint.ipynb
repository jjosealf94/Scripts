```python
%pylab inline
import matplotlib.pyplot as plt
from scipy import misc
from scipy.fftpack import ifft, fft, fftfreq
```

```python
#Lena Normal
lena = misc.lena() 

#Lena negativo
negaLena = zeros((len(lena),len(lena))) 
for i in range(len(lena)): 
    for j in range(len(lena)):
        negaLena[i][j] = 255 - lena[i][j]
        
#Lena volteada horizontal
horiLena =zeros((len(lena),len(lena))) 
for i in range(len(lena)): 
    for j in range(len(lena)):
        horiLena[i][-j] =  lena[i][j]

#Lena volteada vertical
vertiLena =zeros((len(lena),len(lena))) 
for i in range(len(lena)): 
    for j in range(len(lena)):
        vertiLena[-i][j] =  lena[i][j]


```

```python
fig=figure(figsize(8,8))
ax1 =subplot2grid((2,2),(0,0))
ax2 =subplot2grid((2,2),(0,1))
ax3 =subplot2grid((2,2),(1,0))
ax4 =subplot2grid((2,2),(1,1))
ax1.imshow(lena,cmap='gray')
ax2.imshow(negaLena,cmap='gray')
ax3.imshow(horiLena,cmap='gray')
ax4.imshow(vertiLena,cmap='gray')
ax1.axis('off')
ax2.axis('off')
ax3.axis('off')
ax4.axis('off')
fig.tight_layout()
show()
```
