```python
%pylab inline 
from scipy.optimize import curve_fit
```

```python
#Colocando en unidades internacionales
x=[0.023,0.028,0.032,0.037,0.042]
y=[0.034745,0.019689,0.012594,0.007982,0.005822]

def campo(r,a):
    return a/(r**3)

nonlfit=curve_fit(campo,x,y)
momento=nonlfit[0]/(2*10**-7)
print ("El momento dipolar magn√©tico es:%.4f"%momento,"A*m^2")

rlin=linspace(0.023,0.042,100)
figure()
plot(rlin,campo(rlin,nonlfit[0]),"r")
scatter(x,y,label='Datos')
title('Campo magn√©tico de un dipolo')
xlabel('r(m)')
ylabel('B(T)')
legend()
show()

```

```python
Binterpol=campo(rlin,nonlfit[0])

print('Distancia(cm)    Campo B(uT)')
for i in range(len(rlin)):
    print('%.4f            %.2f'%(rlin[i]*10**2,Binterpol[i]*10**6))
```
